```python
import numpy as np
from IPython.display import display, Math



def matrix_rank(mat):
    return np.linalg.matrix_rank(mat)


def has_complement_property(frame, d, names):
    m = len(frame)
    F = np.column_stack(frame)

    total_subsets = 2**m
    for mask in range(1, total_subsets - 1):
        S_idx = []
        C_idx = []
        for j in range(m):
            if (mask >> j) & 1:
                S_idx.append(j)
            else:
                C_idx.append(j)

        S = np.column_stack([F[:, i] for i in S_idx])
        C = np.column_stack([F[:, i] for i in C_idx])

        rS = matrix_rank(S)
        rC = matrix_rank(C)

        if rS < d and rC < d:
            print(" Complement property fails!")
            subset_latex = ", ".join(names[i] for i in S_idx)
            comp_latex   = ", ".join(names[i] for i in C_idx)
            display(Math(r"{Subset: }\{" + subset_latex + r"\}\quad (\text{rank }" + str(rS) + r")"))
            display(Math(r"{Complement: }\{" + comp_latex + r"\}\quad (\text{rank }" + str(rC) + r")"))
            print()

            return False

    return True

n = int(input("Enter number of frames: "))


p = int(input("Enter number of vectors in each frame: "))

mother_sets = []


for k in range(n):
    print(f"\n--- Enter vectors for Frame {k+1} ---")
    current_set = []
    for i in range(p):
        print("Enter vector")
        display(Math(rf"\mathbf{{V}}_{{{k+1}}}^{{{i+1}}}"))
        vec = np.array(list(map(float, input().split())))
        current_set.append(vec)
    mother_sets.append(current_set)


d = len(mother_sets[0][0])

import itertools
index_product = list(itertools.product(*[range(n) for _ in range(p)]))

frame_id = 1
test=0
for choice in index_product:
    frame = []
    frame_names = []

    for pos in range(p):
        set_index = choice[pos]
        frame.append(mother_sets[set_index][pos])
        frame_names.append(rf"\mathbf{{V}}_{{{set_index+1}}}^{{{pos+1}}}")


    frame_id += 1

    ok = has_complement_property(frame, d, frame_names)
    if ok:
      test=test+1
if (test==n**p):
  display(Math(r"\textbf{The complementary property is satisfied for }"
             + rf"\mathbb{{R}}^{{{d}}}"))
```
